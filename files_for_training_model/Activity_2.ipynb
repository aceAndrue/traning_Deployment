```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
url='https://drive.google.com/file/d/1G-VKgBJk-zjoInVXxmHwXpNoMTXdfmng/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
ds = pd.read_csv(dwn_url)
print(ds)
x = ds.iloc[:, :-1].values
y = ds.iloc[:, :-1].values
```

```python
ds['R&D Spend'].fillna(ds['R&D Spend'].mean(), inplace=True)
ds['Administration'].fillna(ds['Administration'].mean(), inplace=True)
ds['Marketing Spend'].fillna(ds['Marketing Spend'].mean(), inplace=True)
ds
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plt.subplot(2,2,1)
x1=ds["R&D Spend"]
y=ds["Profit"]
plt.scatter(x1,y,c='r')
plt.grid()



plt.subplot(2,2,2)
x2=ds["Administration"]
y=ds["Profit"]
plt.scatter(x2,y,c='g')
plt.grid()



plt.subplot(2,2,3)
x3=ds["Marketing Spend"]
y=ds["Profit"]
plt.scatter(x3,y,c='b')
plt.grid()


plt.subplot(2,2,4)
x4=ds["City"]
y=ds["Profit"]
plt.scatter(x4,y,c='y')
plt.grid()


```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder= LabelEncoder()
ds['City'] =label_encoder.fit_transform(ds['City'])
x
```

```python
x = ds[['R&D Spend', 'Marketing Spend']].values
x
```

```python
y = ds['Profit'].values
y
```

```python
from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
x= st_x.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.3, random_state=0)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)

predictions = model.predict(x_test)
np.set_printoptions(precision=2)

len(predictions)

type(predictions)

predictions
```

```python
dataset = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
print(dataset)
```

```python
plt.scatter(x_train[:,0], y_train, color = 'blue')
plt.scatter(x_train[:,1], y_train, color = 'green')
plt.plot(x_test, predictions, color = 'pink')
```
